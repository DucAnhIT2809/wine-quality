# Chia train test

```python
X_r = df_red[['volatile acidity','citric acid','residual sugar','chlorides']]
Y_r = df_red['quality']


x_tr_r, x_te_r = train_test_split(X_r,test_size=0.1,random_state=42)
y_tr_r , y_te_r = train_test_split(Y_r,test_size=0.1,random_state=42)


```

Train Data


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_tr_r,y_tr_r)
y_predicsion = regressor.predict(x_te_r)
y_predicsion = np.round(y_predicsion)
```

```python
from sklearn.metrics import mean_squared_error
RMSE =mean_squared_error(y_te_r,y_predicsion)
print(RMSE)
```
